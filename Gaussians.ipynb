# Idee: Addition von zwei Gauß-Verteilungen

Erster Gauß statisch in der Mitte einer Abmessung $d$, hier $d$ = 1

\begin{align} f_1(x) = \frac{1}{\sqrt{2\pi\sigma}}\exp{\left(-\frac{(x-0.5)²}{2\sigma²}\right)} \end{align}

und ein zweiter Gauß, dessen Position sich durch $a$ dynamisch variieren lässt

\begin{align} f_2(x) = \frac{1}{\sqrt{2\pi\sigma}}\exp{\left(-\frac{(x-a)²}{2\sigma²}\right)} \end{align}

dann die Summe daraus und normaliseren, also

\begin{align} f = \frac{f_1 + f_2}{\text{max}(f_1, f_2)} \end{align}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division

# Init
sigma = 0.35; # >0

# Slider value
a = 0; # 0<a<1

# Variable
x = np.arange(101);
x = x/100;

for index in range(0,10):
    # Move slider
    a = index/10;
    
    # Static Gaussian at 1/2
    f1  = 1/np.sqrt(2*np.pi*sigma)*np.exp(-((x-1/2)**2)/(2*sigma**2));

    # Dynamic Gaussian at a
    f2  = 1/np.sqrt(2*np.pi*sigma)*np.exp(-((x-a)**2)/(2*sigma**2));
    
    # Sum of Gaussians
    fsum = f1+f2;
    
    # Normalize
    fsum = fsum/max(fsum);
    
    # Plot
    plt.plot(x,fsum)
    
    
```
